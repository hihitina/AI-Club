```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Get a few cat and dog filenames
train_cats_dir = '/tmp/cats_and_dogs_filtered/train/cats'
train_dogs_dir = '/tmp/cats_and_dogs_filtered/train/dogs'

cat_fnames = [os.path.join(train_cats_dir, fname) for fname in os.listdir(train_cats_dir)[:5]]
dog_fnames = [os.path.join(train_dogs_dir, fname) for fname in os.listdir(train_dogs_dir)[:5]]

# Display the images
nrows = 2
ncols = 5
plt.figure(figsize=(12, 4))

for i, img_path in enumerate(cat_fnames + dog_fnames):
    sp = plt.subplot(nrows, ncols, i + 1)
    sp.axis('off')

    img = mpimg.imread(img_path)
    plt.imshow(img)

    # Get the folder name ('cats' or 'dogs') and use as label
    label = os.path.basename(os.path.dirname(img_path))
    plt.title(label.capitalize())

plt.show()
```
